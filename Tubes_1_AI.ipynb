# Prep

```python
from random import randint
import numpy as np
import copy
import time
```

```python
n = 5
magicnum = n*(n**3+1)//2
```

```python
def makecube():
  ret = [[[0 for i in range (n)] for j in range(n)] for k in range (n)]
  num = [i+1 for i in range (n*n*n)]
  for i in range (n):
    for j in range (n):
      for k in range (n):
        x = randint(0,len(num)-1)
        ret[i][j][k] = num[x]
        num.pop(x)
  return ret
```

```python
def sumcol(x):
  ans = 0
  for i in range(n):
    for j in range(n):
      temp = 0
      for k in range(n):
        temp += x[i][k][j]
      ans += (temp-magicnum)**2
  return ans

def sumrow(x):
  ans = 0
  for i in range(n):
    for j in range(n):
      temp = 0
      for k in range(n):
        temp += x[k][j][i]
      ans += (temp-magicnum)**2
  return ans

def sumheight(x):
  ans = 0
  for i in range(n):
    for j in range(n):
      temp = 0
      for k in range(n):
        temp += x[i][j][k]
      ans += (temp-magicnum)**2
  return ans

def sumdiag3(x):
  ans = 0
  temp = 0
  for i in range(n):
    temp += x[i][i][i]
  ans += (temp-magicnum)**2
  temp = 0
  for i in range(n):
    temp += x[i][i][n-i-1]
  ans += (temp-magicnum)**2
  temp = 0
  for i in range(n):
    temp += x[i][n-i-1][i]
  ans += (temp-magicnum)**2
  temp = 0
  for i in range(n):
    temp += x[n-i-1][i][i]
  ans += (temp-magicnum)**2
  return ans

def sumdiag2(x):
  ans = 0
  for i in range(n):
    temp = 0
    for j in range(n):
      temp += x[i][j][j]
    ans += (temp-magicnum)**2
    temp = 0
    for j in range(n):
      temp += x[i][j][n-j-1]
    ans += (temp-magicnum)**2
  for i in range(n):
    temp = 0
    for j in range(n):
      temp += x[j][i][j]
    ans += (temp-magicnum)**2
    temp = 0
    for j in range(n):
      temp += x[j][i][n-j-1]
    ans += (temp-magicnum)**2
  for i in range(n):
    temp = 0
    for j in range(n):
      temp += x[j][j][i]
    ans += (temp-magicnum)**2
    temp = 0
    for j in range(n):
      temp += x[n-j-1][j][i]
    ans += (temp-magicnum)**2
  return ans

def score(arr):
  return sumcol(arr) + sumrow(arr) + sumheight(arr) + sumdiag3(arr) + sumdiag2(arr)
```

```python
def makecell(x): #Find the coordinate of x inside the cube
  return x//(n**2), (x%(n**2))//n, x%n

def swap(x,y,arr):
  a, b, c = makecell(x)
  d, e, f = makecell(y)
  ret = copy.deepcopy(arr)
  temp = ret[a][b][c]
  ret[a][b][c] = ret[d][e][f]
  ret[d][e][f] = temp
  return ret

def generate_neighbor(x):
  a = randint(0,n**3-1)
  b = randint(0,n**3-1)
  return swap(a,b,x)
```

# Genetic Algorithm

```python
n_gen = 2**11
p = 0.0001
itr = 300
```

```python
arr = [i for i in range(n**3)]
ind = 0
cb = [[[0 for i in range(n)] for j in range(n)] for k in range(n)]
for i in range(n):
  for j in range(n):
    for k in range(n):
      cb[i][j][k] = arr[ind]+1
      ind += 1
mx = score(cb)
```

```python
def binary_search(p, arr, hi, lo):
  if hi < lo:
    return hi
  mid = (hi+lo)//2
  if p < arr[mid]:
    return binary_search(p, arr, mid, lo)
  elif p >= arr[mid]:
    if p < arr[mid+1]:
      return mid
    else:
      return binary_search(p, arr, hi, mid+1)

def findmax(arr):
  mx = score(arr[0])
  idx = 0
  for i in range(len(arr)):
    if score(arr[i]) > mx:
      mx = score(arr[i])
  return mx

def selection(arr):
  total_score = 0
  partition = [0]
  new_gen = []
  mx = findmax(arr)
  for i in range(len(arr)):
    total_score += (mx - score(arr[i]))
    partition.append(total_score)
  for i in range(n_gen):
    p = randint(0,total_score-1)
    idx = binary_search(p, partition, len(partition)-1, 0)
    new_gen.append(arr[idx])
  return new_gen

def make3d(arr):
  ret = [[[0 for i in range (n)] for j in range(n)] for k in range (n)]
  for i in range(len(arr)):
    a, b, c = makecell(i)
    ret[a][b][c] = arr[i]
  return ret

def cross(arr1, arr2, flags):
  child1 = []
  child2 = []
  temp1 = []
  temp2 = []
  ind = 0
  for i in range(len(flags)):
    a, b, c = makecell(i)
    if flags[i] == 1:
      child1.append(arr1[a][b][c])
      child2.append(arr2[a][b][c])
    else:
      temp1.append(arr1[a][b][c])
      temp2.append(arr2[a][b][c])
      child1.append(temp2[ind])
      child2.append(temp1[ind])
      ind += 1
  child1 = make3d(child1)
  child2 = make3d(child2)
  return child1, child2


def crossover(arr):
  new_gen = []
  for i in range(0,len(arr),2):
    flag = [0 for k in range(n**3)]
    for j in range(n**3):
      if np.random.random() > 0.5:
        flag[j] = 1
    child1, child2 = cross(arr[i], arr[i+1], flag)
    new_gen.append(child1)
    new_gen.append(child2)
  return new_gen

def mutation(arr):
  for i in range(len(arr)):
    if np.random.random() < p:
      a = randint(0,n**3-1)
      b = randint(0,n**3-1)
      arr[i] = swap(a,b,arr[i])
```

```python
gen = [makecube() for i in range(n_gen)]

start = time.time()

for i in range(itr):
  gen = selection(gen)
  gen = crossover(gen)
  mutation(gen)
  print(i)

mins = score(gen[0])
idx = 0
for i in range(len(gen)):
  if score(gen[i]) < mins:
    mins = score(gen[i])
    idx = i

end = time.time()

print("Cube = ", gen[idx])
print("Iteration = ", itr)
print("Score = ", mins)
print("Time = ", end-start,"s")
```
